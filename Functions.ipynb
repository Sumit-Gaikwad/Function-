8.What is a lambda fuction in Python and when is it typically used?

10. What is the difference between map(), reduce(), and filter() function in Python?
* Here's a quick breakdown  of the difference between ```map()```,```reduce()```,```filter()``` in Python:

 1)map()
* Purpose: Applies a function to each item in an iterable and returns as iterable with the results.
* Output:An iterable of transformed values(e.g, list,set).
* Use case:When you want to apply a function to every item in a sequence.
```
number = [1, 2, 3]
squared = map(lambda x: x**2, number) #[1, 4, 9]
```
2)reduce()
* Purpose:Applies a function cumulatively to the items of an iterable, reducing it to a single value.
* Output:A single value.
* Use case:When you want to combine all elements in a sequence to a single result (e.g., sum,product).

3)filter()
* Purpose:Filters items from an iterable based on a function that returns a boolean value.
* Output:An iterable of items that satisty the condition.
* Use case:When you want to select elements based on a condition.
```
numbers = [1, 2, 3, 4]
even = filter(lambda x: % 2 ==0, numbers) # [2, 4]
```
Summary:
* map() transform each item.
* reduce() reduces all items to a single value.
* filter() filters items based on a conditon.

````python
# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
  def __init__(self, n_terms):
    self.n_terms = n_terms
    self.a, self.b = 0, 1
    self.count = 0

  def __iter__(self): # Corrected indentation
    return self

  def __next__(self): # Corrected indentation
    if self.count >= self.n_terms:
      raise StopIteration
    if self.count == 0:
      self.count += 1
      return self.a
    elif self.count == 1:
      self.count += 1
      return self.b
    else:
      self.a, self.b = self.b, self.a + self.b
      self.count += 1
      return self.b

# Example usage:
fib = FibonacciIterator(10)
print(list(fib))
````

````python
# Write a generator functionn in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exponent):
  for i in range(exponent + 1):
    yield 2 ** i
# Example usege:
for power in powers_of_two(5):
  print(power)
````

````python
# Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
  try:
    with open(file_path, 'r', encoding='utf-8') as file:
      for line in file:
        yield line.strip() # Remove trailing newline characters

  except FileNotFoundError:
     print(f"File not found: {file_path}")

# Example usege:
for line in read_file_line_by_line("example.txt"):
  print(line)
````

````python
# Create a Python program that uses 'filter()' to remove all the vowels from a given string.
def remove_vowels(s):
  vowels = "aeiouAEIOU"
  return"".join(filter(lambda char: char not in vowels, s))

#Example usege:
text = "Hello, World!"
result = remove_vowels(text)
print(result)
````
